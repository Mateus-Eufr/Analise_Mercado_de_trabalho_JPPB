ANÁLISE ESTATÍSTICA DO MERCADO DE TRABALHO DE JOÃO PESSOA EM 2024

```python
#importação de bibibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importando base de dados
dados_brutos = pd.read_excel('trabalhojp.xlsx', sheet_name='Planilha1')
display(dados_brutos)
```

```python
#preenchendo espaços em branco por 0
dados_brutos = dados_brutos.replace('(Em branco)', 0)
display(dados_brutos)
```

```python
#melhorando a tabela
caged = dados_brutos.rename(columns={'CNAE 2.0 Divisão':'Subsetores','Vr. Relativa':'Variação Estoque'})
caged = caged.drop(['Grupamento', 'CNAE 2.0 Seção'], axis=1)
caged['Variação Estoque'] = caged['Variação Estoque'] * 100
display(caged)
```

```python
#criação da coluna de participação
estoque = sum(caged['Estoque Mensal'])
caged['Participação'] = caged['Estoque Mensal'].apply(lambda x: (x/estoque) * 100)
display(caged)
```

PRINCIPAIS ESTATÍSTICAS

```python
print('PRINCIPAIS ESTATÍSTICAS')
display(caged.describe())
print(f'Estoque total de empregos em 2024: {estoque}')
```

QUAIS SUBSETORES POSSUEM AS MAIORES PARTICIPAÇÕES NO ESTOQUE DE EMPREGOS?

Podemos observar que 3/5 dos subsetores que detém as maiores participações são do setor de serviços. Com destaque, o subsetor de Informação, comunicação e atividades financeiras (26,67%), que representa serca de 1/4 de todo o emprego formal do munícipio.

```python
print('SUBSETORES COM MAIOR PARTICIPAÇÃO NO ESTOQUE:')
maior_part = caged[caged['Participação'] > 5.95]
maior_part.sort_values(by='Participação', ascending=False)
```

QUAIS SUBSETORES POSSUEM AS MENORES PARTICIPAÇÕES NO ESTOQUE DE EMPREGOS?

Destaque para os subsetores da agropecuária, que possuem as menores participações no estoque de emprego formal. Gerada, principalmente, pela pequena área rural do município.

```python
print('SUBSETORES COM MENOR PARTICIPAÇÃO NO ESTOQUE:')
menor_part = caged[caged['Participação'] < 0.23]
menor_part.sort_values(by='Participação', ascending=True)
```

QUAIS SUBSETORES OBTIVERAM VARIAÇÃO NEGATIVA NO ESTOQUE, COM RELAÇÃO A 2023?

Somente 2 subsetores tiveram variação negativa em 2024. São eles: Indústrias Extrativas e Serviços domésticos.

```python
print('QUAIS SUBSETORES TIVERAM VARIAÇÃO NEGATIVA NO ESTOQUE, COM RELAÇÃO A 2023?')
variacao_negativa = caged[caged['Variação Estoque'] < 0]
variacao_negativa.sort_values(by='Variação Estoque', ascending=True)
```

QUEIS SUBSETORES TIVERAM A MAIOR VARIAÇÃO NO ESTOQUE DE EMPREGOS?

3/5 dos subsetores que tiveram as maiores variações no estoque de emprego formal foram do setor da Contrução Civil. 

Obras de Infra-Estrutura (17,33%);

Serviços Especializados para Construção (14,21);

Construção de Edifícios (10,54%).

```python
print('SUBSETORES COM MAIOR VARIAÇÃO:')
maior_variacao = caged[caged['Variação Estoque'] > 7.92]
maior_variacao.sort_values(by='Variação Estoque', ascending=False)
```

QUAIS SUBSETORES MAIS CONTRATARAM FUNCIONÁRIOS?

Destacam-se os subsetores do setor de serviços (3/5) entre os quais mais admitem funcionários.

```python
print('SUBSETORES QUE MAIS ADMITIDOS (Acima de 75%)')
mais_contratos = caged[caged['Admitidos'] > 6408.7]
mais_contratos.sort_values(by='Admitidos', ascending=False)
```

QUAIS SUBSETORES MAIS DEMITIRAM FUNCIONÁRIOS?

Os mesmos subsetores que mais admitem são os que mais desligam.

```python
print('SUBSETORES QUE MAIS DESLIGARAM (Acima de 75%)')
mais_desligados = caged[caged['Desligados'] > 5950.25]
mais_desligados.sort_values(by='Desligados', ascending=False)
```

GRÁFICO DA VARIAÇÃO DOS ESTOQUES DE EMPREGO FORMAIS EM COMPARAÇÃO A 2023

```python
var_ordenado = caged.sort_values(by='Variação Estoque', ascending=False)
cores = sns.color_palette('coolwarm', len(var_ordenado))
plt.figure(figsize=(12, 8))
barras = plt.barh(var_ordenado['Subsetores'], var_ordenado['Variação Estoque'], color=cores)
plt.title('Variação nos Estoques de Emprego por Subsetor: João Pessoa (2024)', fontsize=14, fontweight='bold')
plt.xlabel('Variação (%)')
plt.grid(axis='x', linestyle='--')
for i, v in enumerate(var_ordenado['Variação Estoque']):
    if v >= 0:
        plt.text(v + 0.5, i, f'{v:.2f}%', va='center', ha='left', fontweight='bold')
    else:
        plt.text(v - 0.5, i, f'{v:.2f}%', va='center', ha='right', fontweight='bold')
plt.xlim(min(var_ordenado['Variação Estoque']) - 10, max(var_ordenado['Variação Estoque']) + 5)
sns.despine()
plt.tight_layout()
plt.savefig('variacao.png')
plt.show()
```

GRÁFICO DA PARTICIPAÇÃO DE CADA SUBSETOR NO ESTOQUE TOTAL DE EMPREGOS FORMAIS

```python
par_ordenado = caged.sort_values(by='Participação', ascending=False)
plt.figure(figsize=(12, 8))
barras_par = plt.barh(par_ordenado['Subsetores'], par_ordenado['Participação'], color=cores)
plt.title('Participação nos Estoques de Emprego Formal por Subsetor: João Pessoa (2024)', fontsize=13, fontweight='bold')
plt.xlabel('Variação (%)')
plt.grid(axis='x', linestyle='--')
for i, v in enumerate(par_ordenado['Participação']):
    plt.text(v + 0.5, i, f'{v:.2f}%', va='center', ha='left', fontweight='bold')
plt.xlim(min(par_ordenado['Participação']), max(par_ordenado['Participação']) + 5)
sns.despine()
plt.tight_layout()
plt.savefig('participacao.png')
plt.show()
```

GRÁFICO DO SALDO DE EMPREGO FORMAIS

```python
saldo_ordenado = caged.sort_values(by='Saldo', ascending=False)
plt.figure(figsize=(12, 8))
barras_saldo = plt.barh(saldo_ordenado['Subsetores'], saldo_ordenado['Saldo'], color=cores)
plt.title('Saldo de Emprego Formal por Subsetor: João Pessoa (2024)', fontsize=13, fontweight='bold')
plt.xlabel('Variação (%)')
plt.grid(axis='x', linestyle='--', alpha=0.6)
for i, v in enumerate(saldo_ordenado['Saldo']):
    if v >= 0:
        plt.text(v + 50, i, f'{v}', va='center', ha='left', fontweight='bold')
    else:
        plt.text(v - 50, i, f'{v}', va='center', ha='right', fontweight='bold')
plt.xlim(min(saldo_ordenado['Saldo'] - 500), max(saldo_ordenado['Saldo']) + 1000)
sns.despine()
plt.tight_layout()
plt.savefig('saldo.png')
plt.show()
```
